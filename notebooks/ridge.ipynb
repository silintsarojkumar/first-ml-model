```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
dataset
```

```python
dataset.info()
```

### Data Cleanning

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
## Remove null value

df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
## Removed  the 122nd row

df =df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix space in coll name
df.columns = df.columns.str.strip()
df.columns
```

```python
## Change the required columns as integer data
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## Changing the other cols to float data types
obj = [features for features in df.columns if df[features].dtype=='object']
```

```python
obj
```

```python
for i in obj:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Lets save the cleanded csv
df.to_csv('Algerian_cleaned_dataset.csv',index=False)
```

## Exploratery Data Analysis

```python
## Drop day , month , year

df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Encoding of the category in classses
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot Density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
## Percenteage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Ploting pie chart
classlabel = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabel,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
```

```python
## Correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```

```python
## Box Plot
sns.boxplot(df_copy['FWI'],color='green')
```

```python
## Monthly fire analysis
dftemp = df.loc[df['Region']==1]
plt.figure(figsize=(20,15))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data = dftemp)
```
